```python
# Question No: 01
# Prime Number Checker
import math
prime_number = int(input("Enter a Number: "))
if prime_number < 2:
    print("Not a Prime Number.")
else:
    for i in range(2, int(math.sqrt(prime_number))):  #math.sqrt(prime_number) returns a float, but range() requires integers.So,Wrap the math.sqrt() in int() or math.isqrt()
        if(prime_number % i == 0):
            print(f"{prime_number} is not a Prime Number.")
            break
        else:
            print(f"{prime_number} is a prime number.")


# Question No: 02
# Palindrome Sentence Checker
import string
sentence4 = input("Enter a sentence: ")
no_punct = ""
for char in sentence4:
    if char not in string.punctuation and char != " ":
        no_punct += char.lower()

if no_punct == no_punct[::-1]:
    print("Palindrome!")
else:
    print("Not a palindrome.")


# Question No: 03
# Simple Calculator ➕➖✖️➗
try:
    num4 = float(input("Enter first number: "))
    operator = input("Enter operator (+, -, *, /): ")
    num5 = float(input("Enter second number: "))

    if operator == "+":
        print(f"Result: {num4 + num5}")
    elif operator == "-":
        print(f"Result: {num4 - num5}")
    elif operator == "*":
        print(f"Result: {num4 * num5}")
    elif operator == "/":
        if num5 != 0:
            print(f"Result: {num4 / num5}")
        else:
            print("❌ Cannot divide by zero.")
    else:
        print("❌ Invalid operator.")
except ValueError:
    print("❌ Please enter valid numbers.")

# Question No: 04
# Guess the Number Game 🎯
import random
secret_number = random.randint(1, 100)
attempts1 = 0

while True:
    try:
        guess = int(input("Guess the number (1-100): "))
        attempts1 += 1
        if guess < secret_number:
            print("Too Low!")
        elif guess > secret_number:
            print("Too High!")
        else:
            print(f"🎉 Correct! You guessed it in {attempts1} attempts.")
            break
    except ValueError:
        print("❌ Please enter a valid number.")

# Question No: 05
# Fibonacci Sequence Generator 🔢
n = int(input("Enter number of terms: "))
first1 = 0
second1 = 1

for i in range(n):
    print(first1, end=" ")
    next_term = first1 + second1
    first1 = second1
    second1 = next_term


# Question No: 06
# Password Validator 🔐
import getpass  # For hiding password input
import string   # For checking special characters

# Securely get the password input from the user
password1 = getpass.getpass("🔐 Enter your password: ")

# Check if password has at least one uppercase letter
has_upper = any(char.isupper() for char in password1)

# Check if password has at least one digit
has_digit = any(char.isdigit() for char in password1)

# Check if password has at least one special character
has_special = any(char in string.punctuation for char in password1)

# Check the length and all required conditions
if len(password1) >= 8 and has_upper and has_digit and has_special:
    print("✅ Strong password!")
else:
    print("❌ Password must be at least 8 characters long and include an uppercase letter, digit, and special character.")


# Question No: 07
# Temperature Converter 🌡️
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15

print(f"🌡️ Fahrenheit: {fahrenheit:.2f}")
print(f"🌡️ Kelvin: {kelvin:.2f}")

# Question No: 08
# Count Vowels and Consonants 🔤
import string

sentence5 = input("Enter a sentence: ")
vowel_count = 0
consonant_count = 0
cleaned = ""

for char in sentence5:
    if char not in string.punctuation and char != " ":
        cleaned += char.lower()

for char in cleaned:
    if char.isalpha():
        if char in "aeiou":
            vowel_count += 1
        else:
            consonant_count += 1

print(f"✅ Vowels: {vowel_count}")
print(f"✅ Consonants: {consonant_count}")

# Question No: 09
# Leap Year Checker 📅
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("✅ Leap Year")
else:
    print("❌ Not a Leap Year")

# Question No: 10
# Shopping Cart Total 💰
total1 = 0
count1 = 0

while True:
    price = input("Enter item price or type 'done' to finish: ")
    if price.lower() == 'done':
        break
    try:
        total1 += float(price)
        count1 += 1
    except ValueError:
        print("❌ Invalid input. Please enter a number.")

if count1 > 0:
    average = total1 / count1
    print(f"🛒 Total items: {count1}")
    print(f"💰 Total price: {total1:.2f}")
    print(f"📊 Average price: {average:.2f}")
else:
    print("No items entered.")
```
